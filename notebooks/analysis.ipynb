## 1. Data Loading

```python
df = pd.read_csv('../data/churn.csv')
print(f'Dataset shape: {df.shape}')
df.head()
```

```python
df.describe()
```

## 2. Churn Distribution

```python
fig, ax = plt.subplots(figsize=(8, 5))
df['Churn'].value_counts().plot(kind='bar', color=['#22c55e', '#ef4444'], ax=ax)
ax.set_title('Churn Distribution', fontsize=14, fontweight='bold')
ax.set_xlabel('Churn')
ax.set_ylabel('Count')
plt.tight_layout()
plt.show()

print(f'Churn Rate: {(df["Churn"] == "Yes").mean():.1%}')
```

## 3. Feature Analysis

Explore key features and their relationship with churn.

```python
# Tenure distribution by churn status
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

for i, churn_val in enumerate(['No', 'Yes']):
    subset = df[df['Churn'] == churn_val]
    axes[i].hist(subset['tenure'], bins=30, color='#8b5cf6' if churn_val == 'No' else '#ef4444', alpha=0.7)
    axes[i].set_title(f'Tenure - Churn={churn_val}', fontweight='bold')
    axes[i].set_xlabel('Tenure (months)')
    axes[i].set_ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
# Contract type vs Churn
pd.crosstab(df['Contract'], df['Churn'], normalize='index').plot(kind='bar', stacked=True, 
    color=['#22c55e', '#ef4444'], figsize=(8, 5))
plt.title('Churn Rate by Contract Type', fontweight='bold')
plt.ylabel('Proportion')
plt.tight_layout()
plt.show()
```
